```python
from sklearn.datasets import load_boston
import sklearn.ensemble
import sklearn.model_selection
import numpy as np

from mindspore_xai.explainer import LIMETabular
```

```python
boston = load_boston()
rf = sklearn.ensemble.RandomForestRegressor(n_estimators=1000)
train, test, labels_train, labels_test = sklearn.model_selection.train_test_split(boston.data, boston.target, train_size=0.80, test_size=0.20)
rf.fit(train, labels_train)
```

```python
categorical_features = list(np.argwhere(np.array([len(set(boston.data[:,x])) for x in range(boston.data.shape[1])]) <= 10).flatten())
feature_names=list(boston.feature_names)
```

```python
train_stats = LIMETabular.to_training_data_stats(train, categorical_features, feature_names)
```

```python
lime = LIMETabular(rf.predict, train_stats, feature_names=feature_names, class_names=['price'], categorical_features=categorical_features)
```

```python
inputs = test[:2]
outputs = lime(inputs, num_features=5)
# outputs format: [[description, weight]]

print('inputs: ', type(inputs), inputs.shape)
print('outputs: ', type(outputs))
```

```python
for i, exps in enumerate(outputs):
    for j, exp in enumerate(exps):
        print('Local explanation for sample'.format(i))
        print(exp, '\n')
```
